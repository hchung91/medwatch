```python
%load_ext autoreload
%autoreload 2

from bs4 import BeautifulSoup
import requests
```

```python
def parse_anchor(anchor):
    """
    Returns the href and content of an anchor

    Parameters:
    -------------
    anchor: bs4 anchor object

    Returns:
    -------------
    href: str - href content found in anchor
    content: str - Content/title found in anchor
    """
    
    hrefs = anchor.get("href")
    content = anchor.text

    return hrefs, content


def find_keywords(anchor, keywords=''):
    """
    Returns keywords found in an anchor
    
    """
    rel_keywords = []
    href, content = parse_anchor(anchor)
    
    for keyword in keywords:
        kw = keyword.lower()
        if kw in href.lower() or kw in content.lower():
            rel_keywords.append(keyword)
    
    return rel_keywords
            
```

```python
r = requests.get('https://www.astrazeneca.com/media-centre/press-releases.html')
page = r.text
page = ' '.join(page.split())
soup = BeautifulSoup(page, 'html.parser')
anchors = list(soup.find('body').find_all('a'))
```

```python
keywords = ['covid-19', 'coronavirus', 'sars-cov-2', 'vaccine']
for anchor in anchors:
    rel_kws = find_keywords(anchor, keywords=keywords)
    print('------------------')
    print(anchor)
    print(rel_kws)
    print('\n')
```
