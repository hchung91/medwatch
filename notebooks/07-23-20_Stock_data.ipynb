```python
%load_ext autoreload
%autoreload 2

import pandas as pd
import yfinance as yf
from datetime import datetime

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# From tutorial:
# https://towardsdatascience.com/historical-stock-price-data-in-python-a0b6dc826836

tickers_list = ['AZN', 'TKBIF', 'MRNA', 'BNTX', 'GSK', 'PFE', 'SHTDY', 'NVAX']

# Cannot specify the start time, only date
data = yf.download(tickers_list,'2020-7-23', interval = '1m')['Adj Close']

# Print first 5 rows of the data
print(data.head())

# Plot all the close prices
((data.pct_change()+1).cumprod()).plot(figsize=(10, 7))

# Render plot labels
plt.legend()
plt.title("Returns", fontsize=16)
plt.ylabel('Cumulative Returns', fontsize=14)
plt.xlabel('Year', fontsize=14)

# Plot the grid lines
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```
